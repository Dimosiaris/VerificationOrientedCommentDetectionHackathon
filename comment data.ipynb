```python
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X = comments[['len','likes','replies']]
y = comments['ver']
columns = ['len','likes','replies','ver']
```

```python
# Split to training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
traindata = pd.DataFrame(data=np.c_[X_train, y_train], columns=columns)
```

```python
traindata
```

```python
# RBF kernel with gamma 1
model = SVC(kernel='rbf', gamma=0.01, C=1.0)
model.fit(X_train, y_train)
```

```python
1 - accuracy_score(y_train, model.predict(X_train))
```

```python
1 - accuracy_score(y_test, model.predict(X_test))
```
